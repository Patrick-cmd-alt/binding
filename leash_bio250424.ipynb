```python
#converts smiles into fingerprint for all the structures
from rdkit import Chem
from rdkit.Chem import AllChem
import pandas as pd

# Function to calculate fingerprints
def calculate_fingerprint(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol is not None:
        fingerprint = AllChem.GetMorganFingerprintAsBitVect(mol, 3, nBits=1024) # 3 is recommended for machine learning
        # Convert fingerprint to list of integers
        fingerprint_list = list(fingerprint.ToBitString())
        return fingerprint_list
    else:
        return None

# Apply the function to calculate fingerprints
df_encoded['fingerprint'] = df_encoded['molecule_smiles'].apply(calculate_fingerprint)

print(df_encoded)
```

```python
from sklearn.utils import resample

# Separate majority and minority classes
df_majority = df_encoded[df_encoded['binds'] == 0]  # Assuming 'binds' is your target column
df_minority = df_encoded[df_encoded['binds'] == 1]

# Downsample the majority class
df_majority_downsampled = resample(df_majority,
                                   replace=False,    # Sample without replacement
                                   n_samples=len(df_minority),  # Match minority class size
                                   random_state=42)  # Reproducible results

# Combine minority class with downsampled majority class
df_downsampled = pd.concat([df_majority_downsampled, df_minority])

# Check the class distribution
print(df_downsampled['binds'].value_counts())

# Now you can use df_downsampled for training your model
```

```python
# Create new columns for each binary value in the 'fingerprint' column
for i in range(len(df_downsampled['fingerprint'].iloc[0])):
    df_downsampled[f'fingerprint_{i}'] = df_downsampled['fingerprint'].apply(lambda x: x[i])

# Convert binary values to boolean
for i in range(len(df_downsampled['fingerprint'].iloc[0])):
    df_downsampled[f'fingerprint_{i}'] = df_downsampled[f'fingerprint_{i}'].astype(bool)

# Drop the original 'fingerprint' column
df_downsampled.drop(columns=['fingerprint'], inplace=True)
```
